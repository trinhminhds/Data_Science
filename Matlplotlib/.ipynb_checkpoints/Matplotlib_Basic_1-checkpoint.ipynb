## Pyplot API 

```python
x = np.linspace(0,10,1000)
plt.plot(x,np.sin(x), color = "red", linestyle = "dashed", label = "sin(x)");
plt.plot(x,np.cos(x), color = "blue", label = "cos(x)");

plt.title("Sin(x) and Cos(x)\n")
plt.xlabel("Bien x")
plt.ylabel("sin(x)")

#plt.xlim([0,4])
#plt.ylim([-0.7,0.7])
#plt.axis([0,4,-0.7,0.8])
#plt.axis("tight")
plt.axis("equal")


plt.legend();
```

# Object-Otiented API

```python
# 1. Prepare data
x = [1,2,3,4,5]
y = [11,12,13,14,15]

# 2. setup plot
fig,ax = plt.subplots(figsize=(5,5))

# 3. Plot the data
ax.plot(x,y);

ax.set(title = "A simple plot\n",
      xlabel = "x-axis",
      ylabel = "y-axis");
```

```python
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)

```

```python
# OO API

# Prepare New Data
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000*rng.rand(100)

fig,ax = plt.subplots()
img1 = ax.scatter(x,y, s = sizes , c = colors, cmap="viridis", alpha=0.5);

fig.colorbar(img1);
```

## Bar
* vertical
* Horizotal

```python
soft_drink_prices = {"Coke": 10,
                    "Pessi": 12,
                    "Cocacola": 9}

fig, ax = plt.subplots()
ax.bar(list(soft_drink_prices.keys()),list(soft_drink_prices.values()));
ax.set(title="Bach Hoa Xanh", ylabel="Price($)");
```

```python
fig,ax = plt.subplots()
ax.barh(list(soft_drink_prices.keys()),list(soft_drink_prices.values()));
```

## Histogram


```python
# Prepare data
np.random.seed(42)
student_height = np.random.normal(170,10,250)
```

```python
student_height[:10], len(student_height)
```

```python
plt.hist(student_height, bins=30);
```

### Subplots

```python
fig , ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5))

# Plot data to each axis
ax1.plot(x,x/2);
ax1.plot(x,x*2);

ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(soft_drink_prices.keys(),soft_drink_prices.values());
ax4.hist(student_height,bins = 30);
```
