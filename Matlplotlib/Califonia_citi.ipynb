```python
import pandas as pd
# Figbug: AttributeError: module 'matplotlib' has no attribute 'scatter'= (import matplotlib.pyplot as plt)
import matplotlib.pyplot as plt
import numpy as np
```

```python
cities = pd.read_csv("./Data/california_cities.csv")
cities.head()
```

```python
# Rename lat = Vi do , lon = Kinh do
lat,lon = cities["latd"], cities["longd"]
population,area = cities["population_total"],cities["area_land_sq_mi"]


plt.style.use('seaborn-v0_8')
plt.figure(figsize=(10,8))

plt.scatter(lon,lat,
           c=np.log10(population),cmap="summer_r",
           s=area,linewidths=2,alpha=0.5);

plt.axis("equal");
plt.xlabel("longtatitude")
plt.ylabel("lattitude")
plt.colorbar(label = "log$_{10}$(population)")
plt.clim(3,7)

# Create a legend for cities size
area_range = [50,150,300,500]

for area in area_range:
    plt.scatter([],[],s=area,c="k",alpha=0.4,
               label = str(area) + "km$^2$")
    
plt.legend(labelspacing=1,title = "City Area")

# Name chart
plt.title("POPULATION AND AREA DISTRIBUTION CALIFORNIA CITIES\n")
```
