```python
# Conversion Operation on Series

# Reading csv file from url
df = pd.read_csv('https://media.geeksforgeeks.org/wp-content/uploads/nba.csv')

# Dropping null value columns to avoid errors
df.dropna(inplace = True)

# storing dtype before converting
before = df.dtypes

# converting dtypes using astype
df['Salary'] = df['Salary'].astype(int)
df['Number'] = df['Number'].astype(str)

# storing dtype before converting
after = df.dtypes

print('Before Conversion: \n', before)
print('\n\n')
print('After Conversion: \n', after)
```

```python
# Conversion Operation on Series
# using tolist() function

# Reading csv file from url
df1 = pd.read_csv('https://media.geeksforgeeks.org/wp-content/uploads/nba.csv')

# Dropping null value columns to avoid errors
df1.dropna(inplace = True)

# storing dtype before operation
dtype_before = type(df1['Salary'])

# Converting to list
salary_list = df1['Salary'].tolist()

# storing dtype after operation
dtype_after = type(salary_list)

print('Before Conversion: \n', dtype_before)
print('\n\n')
print('After Conversion: \n', dtype_after)
```

## Pandas Series.str.slice()

```python
# In this example, the salary column has sliced to get values before decimal.
# For example, we want to do some mathematical operations and for that we need interger data,
# So the salary columns will be sliced till the 2nd last element (-2 position)

# Since the salary column is imported as float64 data type, it is first converted to string using the
# .astype() menthod

df = pd.read_csv('https://media.geeksforgeeks.org/wp-content/uploads/nba.csv' , index_col='Name')

df.dropna(inplace = True)

df['Salary'] = df['Salary'].astype(str)

df['Salary (int)'] = df['Salary'].str.slice(0, -2, 1)

df.head(10)
```

## Pandas.apply()

```python
# The following example passes a function and checks the values of each element
# in series and returns low, normal or high accordingly

df  = pd.read_csv('/content/nba.csv')

def func(num):
  if num < 200:
    return 'Low'
  elif num >= 200 and num < 400:
    return 'Normal'
  else:
    return 'High'

new = df['Weight'].apply(func)

new.head(10)
```

```python
# In the following example, a temporary anonyramous function is made in .apply itself using lambla.
# It adds 5 to each value in series and returns a new series

df  = pd.read_csv('/content/nba.csv')

print(df['Weight'][:10])

df['Weight'] = df['Weight'].apply(lambda x: x + 5)

print(df['Weight'][:10])
```

## Apply function to every row in a Pandas DataFrame

```python
# Applying User-Defined Function to Every Row of Pandas DataFrame

def add_values(row):
  return row['A'] + row['B'] + row['C']

def main():
  data = {
        'A': [1, 2, 3],
        'B': [4, 5, 6],
        'C': [7, 8, 9]}

  df = pd.DataFrame(data)

  print('Original Data: \n', df)

  df['add'] = df.apply(add_values, axis = 1)

  print('After Applying Function: \n',df)

if __name__ == '__main__':
  main()
```

```python
# Apply Lambda to Every Row of Dataframe

def add(a, b, c):
  return a + b + c

def main():

  data = {
        'A': [1, 2, 3],
        'B': [4, 5, 6],
        'C': [7, 8, 9]}

  df = pd.DataFrame(data)

  print('Original Data: \n', df)

  df['add'] = df.apply(lambda x: add(x['A'], x['B'], x['C']), axis = 1)

  print('After Applying Function: \n',df)

if __name__ == '__main__':
  main()
```

## Pandas Serias.apply()

```python
# Pandas Series.apply() function invoke the passed function on each element of the given series object.

sr = pd.Series(['New York', 'London', 'Paris', 'Rio', 'Tokyo'])

index_ = ['City 1', 'City 2', 'City 3', 'City 4', 'City 5']

sr.index = index_

sr
```

```python
# Use Series.apply() function to change the city name to 'Motreadl' if the city is 'Rio'

result = sr.apply(lambda x: 'Motreadl' if x == 'Rio' else x)

result
```

```python
# Use Series.apply() function to return True if the value in the given series
# object is greater than 30 else return False

sr = pd.Series([10, 20, 30, 40, 50])

index_ = pd.date_range('2021-01-01', periods = 5, freq = 'Y')

sr.index = index_

print(sr)

result = sr.apply(lambda x: True if x > 30 else False)

print('\n',result)
```

## Pandas DataFrame.truncate()

```python
# Use DataFrame.truncate() function to truncate some entries before and after
# the passed labels of the given dataframe.


df = pd.DataFrame({'Weight':[45, 88, 56, 15, 71],
                   'Name':['Sam', 'Andrea', 'Alex', 'Robin', 'Kia'],
                   'Age':[14, 25, 55, 8, 21]})

index_ = pd.date_range('2010-10-09 08:45', periods = 5, freq ='H')

df.index = index_

df
```

```python
# Now we will use DataFrame.truncate() function to truncate
# the entries before ‘2010-10-09 09:45:00’ and after ‘2010-10-09 11:45:00’ in the given dataframe.

result = df.truncate(before = '2010-10-09 09:45:00', after = '2010-10-09 11:45:00')

result
```
